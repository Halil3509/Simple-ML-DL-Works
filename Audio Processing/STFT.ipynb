STFT (Short-time Fourier Transform)  zamanla değişen bir sinyalin yerel bölümlerinin sinozoidal frekansını ve faz içeriğini belirlemek için kullanılan Fourier ile ilgili bir dönüşümdür. Uygulamada STFT'leri hesaplama prosedürü, daha uzun bir zaman sinyalini eşit uzunlukta daha kısa parçalara bölme ve ardından Fourier dönüşümünü her kısa parça üzerinde ayrı ayrı hesaplamaktır.

* Her kısa segmentte Fourier spektrumu ortaya çıkarır.
* Sonra da spectrogram olarak bilinen dğeişen spekturumları zamanın bir fonksiyonu olarak çizer. 

## Continuous-time STFT

Devamlı sinyallerde STFT işlemi formülde de görüldüğü üzere integral alınarak yapılır.

## Discrete-time STFT
